# Galaxy vs Quasars

## Generative Classification

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.visualization import hist
import pandas as pd
import corner
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('../solutions/galaxyquasar.csv', delimiter=',')
data['u-g'] = data['u'] - data['g']
data['g-r'] = data['g'] - data['r']
data['r-i'] = data['r'] - data['i']
data['i-z'] = data['i'] - data['z']
data['class-int'] = np.where(data['class']=='GALAXY', 0, 1)
#print(data['class'])
#print(data['class-int'])
```

```python
labels = ['u-g', 'g-r', 'r-i', 'i-z']
#Assign to variable otherwise plot shows twice
corner = corner.corner( np.array([data['u-g'],data['g-r'],data['r-i'],data['i-z']]).T, labels=labels)
```

```python
hist(data['u-g'], bins='freedman', histtype = 'step', label = 'All data')
hist(data['u-g'][data['class']=='GALAXY'], bins='freedman', histtype = 'step', label = 'Galaxy')
hist(data['u-g'][data['class']=='QSO'], bins='freedman', histtype = 'step', label = 'QSO')
plt.legend()
plt.grid()
```

### Prepare Data 

```python
from sklearn.model_selection import train_test_split
from astroML.utils import completeness_contamination
from sklearn.metrics import roc_curve
```

```python
X = np.array([data['u-g'],data['g-r'],data['r-i'],data['i-z']]).T
y = np.array(data['class-int'])  
print(X.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=12)
print(X_train.shape)
```

### Gaussian Naive Bayes

#### ROC Curve

```python
from sklearn.naive_bayes import GaussianNB

n_colors = 4

y_pred = []
y_prob = []

for i in range(n_colors):
    X_tr = X_train[:,0:i+1]
    X_te = X_test[:,0:i+1]
    print(X_tr.shape)
    
    gnb_cl = GaussianNB().fit(X_tr, y_train)
    y_pred.append(gnb_cl.predict(X_te))
    y_prob.append(gnb_cl.predict_proba(X_te)[:,1])
    
    #ROC Curve
    fpr, tpr, thresh = roc_curve(y_test, y_prob[-1])
    plt.plot(fpr, tpr, label=labels[0:i+1])

plt.title('ROC')
plt.legend()
plt.xlim(0,0.1)
plt.ylim(0.9,1)
plt.grid()
plt.show()
```

#### Contamination and Completeness

```python
#Contamination and completeness
completeness, contamination = completeness_contamination(y_pred, y_test)

fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(range(1,n_colors+1), contamination, marker='o', label ='contamination')
ax[1].plot(range(1,n_colors+1), completeness, marker='o', label = 'completeness')

ax[1].set_xlabel(r'$N_{colors}$')

ax[0].set_ylabel('Contamination')
ax[0].grid(True)

ax[1].set_ylabel('Completeness')
ax[1].grid(True)
plt.tight_layout
```

#### Result

```python
X_tr = X_train[:,0:1] #n_colors=1
gmm_cl = GaussianNB().fit(X_tr, y_train)
y_pred = gmm_cl.predict(X[:,0:1])
```

### Pytorch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data as torchdata

from astroML.utils.decorators import pickle_results
```

```python
print(X.shape)
```

```python
datanormed = np.zeros((len(X), 6), 
                      dtype=np.float32)
for i in range(X.shape[1]):
    datanormed[:,i] = X[:,i]
datanormed[:,5] = y.astype('float32')
```

```python
class Net(nn.Module):
    def __init__(self, nhidden):
        super(Net, self).__init__()
        self.fc_h = nn.Linear(5, nhidden)
        self.fc_o = nn.Linear(nhidden, 1)

    def forward(self, x):
        h = F.relu(self.fc_h(x))
        z = self.fc_o(h)
        return z
```

```python
# split data into 7:3 train:test
dataset = torchdata.TensorDataset(torch.tensor(datanormed[:, 0:5]),
                                  torch.tensor(datanormed[:, 5]).view(-1, 1))
trainnum = datanormed.shape[0] // 10 * 7
traindata, testdata = torchdata.random_split(dataset, 
                                             [trainnum, datanormed.shape[0] - trainnum])
traindataloader = torchdata.DataLoader(traindata, 
                                       batch_size=128, 
                                       shuffle=True)
```

```python
@pickle_results('NNquasar-gal.pkl')
def train_NN():
    # 4 hidden layers
    model = Net(4)
    # MSE loss
    criterion = torch.nn.MSELoss(reduction='sum') 
    # stochastic gradient descent
    optimizer = torch.optim.SGD(model.parameters(), lr=0.001) 
    # Adaptive learning rate...
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, 
                                                           verbose=True, 
                                                           patience=5, 
                                                           threshold=1e-3)

    min_valid_loss = float('inf')
    badepochs = 0
    for t in range(100):
        train_loss = 0
        for i, databatch in enumerate(traindataloader, 0):
            photometry, redshifts = databatch
            optimizer.zero_grad()
            z_pred = model(photometry)
            loss = criterion(z_pred, redshifts)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        with torch.no_grad():
            photometry = testdata[:][0]
            redshifts = testdata[:][1]
            z_pred = model(photometry)
            valid_loss = criterion(z_pred, redshifts)
            if t % 10 == 0:
                print('Epoch %3i: train loss %0.3e validation loss %0.3e' % (t, \
                        train_loss / len(traindata), valid_loss / len(testdata)))
            # stop training if validation loss has not fallen in 10 epochs
            if valid_loss > min_valid_loss*(1-1e-3):
                badepochs += 1
            else:
                min_valid_loss = valid_loss
                badepochs = 0
            if badepochs == 10:
                print('Finished training')
                break
        scheduler.step(valid_loss)
    return model
```

```python
model = train_NN()
```

```python
# plot the results
from sklearn.metrics import accuracy_score
with torch.no_grad():
    X_test = testdata[:][0]
    y_test = testdata[:][1]
    y_pred = model(X_test)

    y_pred = y_pred.numpy().flatten().astype(int)
    y_test = y_test.numpy().flatten().astype(int)

    print(y_pred)
    print(y_test)

    print(accuracy_score(y_test, y_pred))

    # hist(X_test[y_test == 0], bins='freedman', histtype = 'step', label = 'Galaxy')
    # hist(X_test[y_test == 1], bins='freedman', histtype = 'step', label = 'QSO')

    # hist(X_test[y_pred == 0], bins='freedman',label = 'Galaxy Predicted', alpha=0.3)
    # hist(X_test[y_pred == 1], bins='freedman', label = 'QSO Predicted',alpha=0.3)
    
    # plt.legend()
    # plt.grid()
```

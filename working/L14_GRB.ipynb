```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import corner
```

```python


# Download file
r = requests.get('https://user-web.icecube.wisc.edu/~grbweb_public/Summary_table.txt')
with open("Summary_table.txt", 'wb') as f:
    f.write(r.content)

# Read content
data = np.loadtxt("Summary_table.txt", dtype='str',unpack='True')

# Read headers
with open("Summary_table.txt",'r') as f:
    names= np.array([n.strip().replace(" ","_") for n in f.readlines()[1].replace("#","").replace("\n","").lstrip().split('    ') if n.strip()!=''])
```

```python
print(names)
```

```python
#Focus only on these data
t90 = np.array(data[6], dtype=float)#time interval of 90% photons counts
fluence = np.array(data[9], dtype=float) #The time integrated total energy per unit surface
z = np.array(data[11], dtype=float) #redshift

#Noticed outliers on data -> initial mask just to have corner (same # points for all quantities)
mask_t90 = (t90 >0)
mask_fluence = (fluence >0)
mask_z = (z >=0)
mask =  mask_t90 & mask_fluence & mask_z

#!all data are in log10
data_analysis = np.array([np.log10(t90[mask]), np.log10(fluence[mask]), (z[mask])]).reshape(len(t90[mask]), 3)
```

```python
fig = corner.corner(
    data_analysis, labels=['t90', 'fluence', 'z'], levels=[0.68,0.95], quantiles=(0.16, 0.5, 0.84), use_math_text=True, show_titles=True, 
);
```

Given the corner plot, it seems that there exists two populations of GRB that could be identified using either quantity among redshift, fluence and t90. However, the redshift values do not seem right as there are displayed negative values. In the following, I'll plot the distributions separately.

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 5))
ax[0].hist(z[mask_z], bins='fd', density=True, histtype='step')
ax[0].set_xlabel('Redshifts')
ax[0].grid()

ax[1].hist(np.log10(t90[mask_t90]), bins='fd', density=True, histtype='step')
ax[1].set_xlabel('Log T90')
ax[1].grid()

ax[2].hist(np.log10(fluence[mask_fluence]), bins='fd', density=True, histtype='step')
ax[2].set_xlabel('Log Fluence')
ax[2].grid()

plt.tight_layout()
```

By eye, a bimodality could be seen in the distribution of t90, thus I'll analyse that quantity. 

```python
t90 = np.log10(t90[mask_t90])
```

## KDE

```python
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV


xgrid = np.linspace(-2, 3, 1000)

def kde_sklearn(data, bandwidth):
    kde_skl = KernelDensity(bandwidth = bandwidth)
    kde_skl.fit(data[:, np.newaxis])
    log_pdf = kde_skl.score_samples(xgrid[:, np.newaxis]) 
    return np.exp(log_pdf)

# Cross Validation to find the bandwidth
bwrange = np.linspace(0.01,1.0, 50) 
K = 5 
grid = GridSearchCV(KernelDensity(), {'bandwidth': bwrange}, cv= K)
grid.fit(t90[:, None])
best_bw = grid.best_params_['bandwidth']
print(best_bw)
```

```python
pdf = kde_sklearn(t90,bandwidth=best_bw)

plt.plot(xgrid,pdf, label='KDE')

plt.hist(t90, bins='fd', density=True, histtype='step', label='data')
plt.xlabel('Log T90')
plt.legend()
plt.grid()
```

## Gaussian Mixtures

```python
from sklearn.mixture import GaussianMixture

#plt.plot(xgrid,pdf, label='KDE')

plt.hist(t90, bins='fd', density=True, histtype='step', label='data')
plt.xlabel('Log T90')



t90_reshaped = t90.reshape(-1,1)
aic = []
n_comp = np.arange(1,6, 1)
for i in n_comp:
    gauss_mixtures = GaussianMixture(n_components=i).fit(t90_reshaped)
    log_likelihood = gauss_mixtures.score_samples(xgrid.reshape(-1,1))
    likelihood = np.exp(log_likelihood)

    plt.plot(xgrid, likelihood, label = 'N component: {}'.format(i), linewidth = 0.8)
    aic.append(gauss_mixtures.aic(t90_reshaped))
plt.legend(prop={'size': 8})
plt.grid()

print(n_comp, aic)

```

```python
plt.plot(n_comp, aic, marker = 'o',label = 'AIC')
plt.grid()
plt.legend()
plt.xlabel('# Components')
plt.ylabel('Criterion')
```

According to the AIC using Gaussian Mixtures, the number of components that best represent the data is 2.

```python

plt.hist(t90, bins='fd', density=True, histtype='stepfilled', alpha = 0.3,label='data')
plt.xlabel('Log T90')

ind = np.where(aic == min(aic))
best_n = n_comp[ind][0] 

model = GaussianMixture(n_components=best_n).fit(t90_reshaped)
ncomp_density = model.predict_proba(xgrid.reshape(-1,1)) #pdf of the single components
individual_pdf = ncomp_density*likelihood[:,np.newaxis]

for i in range(best_n): 
    plot, = plt.plot(xgrid, individual_pdf[:,i], label = ' GM Component {}'.format(i+1))
    plt.axvline(model.means_[i], color=plot.get_color())
    plt.axvline(model.means_[i] - model.covariances_[i], color=plot.get_color(), linestyle='--')
    plt.axvline(model.means_[i] + model.covariances_[i], color=plot.get_color(), linestyle='--')
    print('Mean : {}, Sigma: {}'.format(model.means_[i], model.covariances_[i]))
plt.legend(prop={'size':8})
plt.grid()
```

The intersection between the two populations marks the threshold between 'short' and 'long' GRB.

```python
mask_grid = xgrid>-1 #not exactly same value, disregard the first points as tails of both distributions
diff = np.abs(individual_pdf[mask_grid][:,0] - individual_pdf[mask_grid][:,1])
ind = np.argmin(diff)
threshold = xgrid[mask_grid][ind]
print(threshold)
```

```python

plt.hist(t90, bins='fd', density=True, histtype='stepfilled', alpha = 0.3,label='data')
plt.xlabel('Log T90')

for i in range(best_n): 
    plot, = plt.plot(xgrid, individual_pdf[:,i], label = ' GM Component {}'.format(i+1))
    #plt.axvline(model.means_[i], color=plot.get_color())
    #plt.axvline(model.means_[i] - model.covariances_[i], color=plot.get_color(), linestyle='--')
    #plt.axvline(model.means_[i] + model.covariances_[i], color=plot.get_color(), linestyle='--')
    #print('Mean : {}, Sigma: {}'.format(model.means_[i], model.covariances_[i]))
plt.axvline(threshold, label='Threshold')
plt.legend(prop={'size':8})
plt.grid()
```

## Clustering (KMeans)

```python
from sklearn.cluster import KMeans

clf = KMeans(n_clusters=2)          
clf.fit(t90_reshaped)
centers = clf.cluster_centers_   
labels = clf.predict(t90_reshaped)          
print(centers)
```

```python
plt.hist(t90, bins='fd', density=True, histtype='stepfilled', alpha = 0.3,label='data')
plt.xlabel('Log T90')

for i in range(best_n): 
    plot, = plt.plot(xgrid, individual_pdf[:,i], label = ' GM Component {}'.format(i+1))
    plt.axvline(model.means_[i], color=plot.get_color())
    #plt.axvline(model.means_[i] - model.covariances_[i], color=plot.get_color(), linestyle='--')
    #plt.axvline(model.means_[i] + model.covariances_[i], color=plot.get_color(), linestyle='--')
    print('Mean : {}, Sigma: {}'.format(model.means_[i], model.covariances_[i]))
plt.legend(prop={'size':8})


for i in range(2):
    plt.axvline(centers[i], label = 'Center KMeans {}'.format(i+1, centers[i]))
    print('Center KMeans: ', centers[i])
plt.grid()
plt.legend()
```

Both KMeans and Gaussian Mixtures find similar mean values accounting for 2 populations.

Having labeled the two populations with the KMeans, the frequency of short and long GRB can be computed.

```python
tot = len(labels)
long = np.sum(labels) /tot
short = 1-long
print('Frequency of short GRB: ',short)
print('Frequency of long GRB: ', long)
```

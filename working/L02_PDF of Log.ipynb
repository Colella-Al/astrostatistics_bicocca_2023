# Probability Density of log

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#Sample data - Normal Distribution
N=100000
x = np.random.uniform(0.1,10,N) 

#Related histogram
plt.hist(x, density=True)
plt.grid()
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Uniform Distribution')
```

The normal distribution has been 'transformed' with a log10. 

As x is a random variable, also f(x) would be a random variable as well.


```python
#Sample data - Log from Normal Distribution
y=np.copy(x)
y=np.log10(y)

#Related Histogram
plt.hist(y,density=True)
plt.grid()
plt.title('Log10 of a normal distribution')
plt.xlabel('y')
plt.ylabel('p(y)')
```

The distribution is trasformed as:
$$p(y)= |\frac{dx}{dy}|p(x)$$
resulting from considering the conservation of **dimensionless probability**: $p(x)dx=p(y)dy$ 



In the case of y=log10(x):
$$|\frac{dy}{dx}| = \frac{1}{x \ln(10)} = \frac{1}{10^{y}\ln(10)}$$ 

```python
#Pdf
def pdf_log(x):
    return (10**(x))*np.log(10)
normalization = integrate.quad(pdf_log, -1, 1)
plt.hist(y, density=True)
plt.scatter(y,(10**(y))*np.log(10)/normalization[0], color='orange', s=0.5) #normalized with the area->quad
plt.grid()
plt.title('Probability density of a log10 normal distribution')
plt.xlabel('y')
plt.ylabel('p(y)')
```

**Mean and Median**
The mean of a log does not correspond to the log of a mean.
Computing the median that is: the middle number in a sorted data. Therefore the log of the median and the median of log are the same.


```python
print('Mean Normal: ', np.mean(x))
print('Log of mean normal: ', np.log10(np.mean(x)))
print('Median Normal: ', np.median(x))
print('Log of median normal: ', np.log10(np.median(x)))
print('Mean Log: ', np.mean(y))
print('Median Log: ', np.median(y))
```
